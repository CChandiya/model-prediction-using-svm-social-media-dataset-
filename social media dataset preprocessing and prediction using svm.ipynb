```python
def convert_to_numeric(value):
    if isinstance(value, str):
        if 'M' in value:
            return float(value.replace('M', '')) * 1e6
        elif 'K' in value:
            return float(value.replace('K', '')) * 1e3
        else:
            return float(value)
    else:
        return float(value)
```

```python
dataset['Subscribers'] = dataset['Subscribers'].apply(convert_to_numeric)
dataset['Views avg.'] = dataset['Views avg.'].apply(convert_to_numeric)
dataset['Likes avg.'] = dataset['Likes avg.'].apply(convert_to_numeric)
dataset['Comments avg.'] = dataset['Comments avg.'].apply(convert_to_numeric)
dataset['Shares avg.'] = dataset['Shares avg.'].apply(convert_to_numeric)
```

```python
dataset['Score'] = dataset['Subscribers'] + dataset['Views avg.'] + dataset['Likes avg.'] + dataset['Comments avg.'] + dataset['Shares avg.']
Average_score = dataset['Score'].mean()
dataset['Status'] = dataset['Score'].apply(lambda x: 'High' if x >= Average_score else 'Low')
```

```python
dataset.head(5)
```

```python
from sklearn import preprocessing 
l=preprocessing.LabelEncoder()
status_encoded=l.fit_transform(dataset['Status'])
dataset['Status']=status_encoded
```

```python
x=dataset.iloc[:,3:8]
```

```python
x
```

```python
y=dataset.Status
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.svm import SVC
classifier=SVC(kernel='linear',random_state=0)
classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score, classification_report
accuracy = accuracy_score(y_test,y_pred)
print(f'Accuracy: {accuracy}')
```
